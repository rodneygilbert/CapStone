```python
import pandas as pd
import sqlite3
# set connection to SQLite Database
conn = sqlite3.connect("MHPOLL.db")
c = conn.cursor()
# open and read file
df = pd.read_csv('mhpoll.csv')
# print(a1)
df.head()
# drop the city column from dataframe
df2 = df.drop(['City'], axis=1)
# rename event category. change to 'Response'
df2.rename(columns={'Event Category': 'Response'}, inplace=True)
df2.head()
print(df2)
# remove nulls from data
df2.dropna()
print(df2)
# export altered csv
df2.to_sql('mhpoll.sql', conn,if_exists='replace', index=False)



```
